```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.columns
```

```python
df['attendance_days'] = df['attendance_days'].fillna(0)
```

```python
df['overtime_hours'] = df['overtime_hours'].fillna(0)
```

```python
df['salary_credit_date'] = pd.to_datetime(df['salary_credit_date'], errors='coerce')
```

```python
df['attendance_missing'] = np.where(df['attendance_days'] == 0, 'Yes', 'No')
```

```python
df['salary_leakage'] = np.where(
    df['salary_credit_amount'] > df['salary'],
    df['salary_credit_amount'] - df['salary'],
    0
)
```

```python
df['overtime_abuse'] = np.where(df['overtime_hours'] > 40, 'Yes', 'No')
```

```python
df['ghost_employee'] = np.where(
    (df['employment_status'] == 'Inactive') & (df['salary_credit_amount'] > 0),
    'Yes',
    'No'
)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['department'] = df['department'].str.strip().str.title()
df['designation'] = df['designation'].str.strip().str.title()
df['employment_status'] = df['employment_status'].str.strip().str.title()
```

```python
df['cost_per_employee'] = df['salary_credit_amount'] + df['overtime_pay']
```

```python
df['payroll_accuracy'] = np.where(
    (df['salary_leakage'] == 0) & (df['attendance_missing'] == 'No'),
    'Accurate',
    'Review'
)
```

```python
df.to_csv("hr_payroll_cleaned.csv", index=False)
```
